```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy
```

```python
np.random.seed(613)
x = np.arange(50) #is giving integers from 0 to 49
y = np.random.randint(0, 100,50) #1st value is our min, 2nd value is our max, 3rd value is how many numbers (is 50 random integers between 0 and 99)
```

```python
fig, ax = plt.subplots(figsize=(5, 3)) #5inch X 3inch
ax.scatter(x,y) #scatter plot
```

```python
fig, ax = plt.subplots(figsize=(5, 3))
ax.bar(x,y) #bar plot
```

```python
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,y)
```

```python
fig, ax = plt.subplots(figsize=(5, 3))
ax.hist(y) # counting the numbers eg. we have 9 from 85
```

```python
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,y)

ax.set_title('Total growth over time')
ax.set_ylabel('Total growth')
ax.set_xlabel('Years since start')
```

```python
font1 = {'family':'sans-serif','color':'blue','size':20}
font2 = {'family':'monospace','color':'green','size':14}

fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,y)

ax.set_title('Total growth over time', fontdict = font1)
ax.set_ylabel('Total growth', fontdict = font2)
ax.set_xlabel('Years since start', fontdict = font2)

```
